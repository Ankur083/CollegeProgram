```python
import numpy as np


even_array = np.arange(10, 10 + 4 * 5 * 2, 2).reshape(4, 5)
third_column = even_array[:, 2]

even_array[3] = [1, 2, 3, 4, 5]
print("Original 4x5 Array of Even Numbers:")
print(even_array)

print("\nThird Column:")
print(third_column)
```

```python
import numpy as np

names = np.array(["Roxana", "Statira", "Roxana", "Statira", "Roxana"])
score = np.array([126, 115, 130, 141, 132])

scores_below_130 = score[score < 130]
scores_statira = score[names == "Statira"]
score[names == "Roxana"] += 10

print("Scores below 130:", scores_below_130)
print("Scores by Statira:", scores_statira)
print("Updated scores:", score)
```
