```python
# Q33 input a list of ten numbers from the user and sort them in ascending order using the bubble sort?
list=[]
for i in range(10):
    num=int(input("enter a number"))
    list.append(num)
print(list)
for k in range(9):
    for j in range (9-k):
        if list[j]>list[j+1]:
            # list[j],list[j+1]=list[j+1],list[j]
           t=list[j]
           list[j]=list[j+1]
           list[j+1]=t
print(list)        
```
