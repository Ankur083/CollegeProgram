```python
def powerfun(a,n):
    prod =1
    for i in range(n):
        prod = prod*a
        
    return prod 
print(powerfun(2,5))
```

```python
def recpower(a,n):
    if n==1:
        return a
    elif (n%2==0):
        return recpower(a,n/2)*recpower(a,n/2) #a^n = a^n/2*a^(n/2)
    else:
        return a*recpower(a,(n-1)/2)*recpower(a,(n-1)/2) #a^n = a*^((n-1/2)*a^((n-1)/2)
print(recpower(2,9))
```
